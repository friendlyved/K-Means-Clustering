```python

def remove_outliers (df,col,k):
    mean = df [col].mean()
    global df1
    sd =df[col].std()
    final_list = [x for x in df[col] if (x > mean - k*sd)]
    final_list = [ x for x in final_list if ( x < mean + k*sd)]
    df1 =df.loc[df[col].isin(final_list)]; print(df1.shape)
    print(" no of outlier removed",df.shape[0]-df1.shape[0])
```

```python
ssd=[]
for k in range(1,11):
    mkt_kmeans=KMeans(n_clusters=k)
    mkt_kmeans.fit(mkt)
    scr=mkt_kmeans.score(mkt)
    scr=np.round(np.abs(scr))
    ssd.append(scr)
numbers=list(range(1,11))
plt.figure(figsize=(6,6))           
plt.plot(numbers,ssd,marker="*",color='r')
plt.grid()
plt.xlabel('No Of Clusters',size=14)
plt.ylabel('SSD',size=14)
plt.title('Elbow Plot For Clustering Algorithm',size=18,color='r',pad=20)
    
```

```python
df_with_label=pd.concat([mkt,pd.Series(mkt_kmeans.labels_)],axis=1)
df_with_label.rename(columns={df_with_label.columns[29]:'Cluster_number'},inplace=True)
df_with_label
```

```python
df_with_label.Cluster_number.value_counts()
```

```python
# see every group's mean value of every columns.
df_with_label_groupby = df_with_label.groupby(by='Cluster_number').mean().reset_index()
df_with_label_groupby
```

```python
len(mkt.columns)
```

```python
# visualize and find out that only AcceptedCmp2 and Response has obvious difference from groups.
# If there is obvious difference between groups, we can explain behavior of each group from the feature.
for index, col in enumerate(df_with_label.columns[2:]):
    fig = plt.figure(figsize=(24,42))
    plt.subplot(24,3,index+1)
    sns.barplot(x='Cluster_number', y=col, data=df_with_label_groupby)
    plt.show()
```

```python
# check if we can see the group seperated obviously.
fig = plt.figure(figsize=(20,10))
ax = plt.axes()
sns.scatterplot(x='MntMeatProducts', y='NumCatalogPurchases', hue='Cluster_number', data=df_with_label)
plt.show()
```

```python
# use PCA to decrease the dimesions to 2, to let us see the group seperated easily.
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
pca_array = pca.fit_transform(mkt)
pca_array
```

```python
# put data in dataframe
df_pca = pd.DataFrame(pca_array, columns=['x', 'y'])
df_pca['Cluster_number'] = mkt_kmeans.labels_
df_pca
```

```python
# check if we can see the group seperated obviously. 
fig = plt.figure(figsize=(20,10))
ax = plt.axes()
sns.scatterplot(x='x', y='y', hue='Cluster_number', data=df_pca)
plt.show()
```
